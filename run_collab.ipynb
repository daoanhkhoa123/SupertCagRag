```python
#
"""Run script for the Streamlit application."""
import subprocess
import sys
from pathlib import Path
import logging

from pyngrok import ngrok

# Configure logging
PORT = 8000
logging.basicConfig(filename='chainlit.log', level=logging.DEBUG,
                    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s')
logger = logging.getLogger(__name__)

def main():
    """Run the Streamlit application."""
    # Add your ngrok authtoken
    !ngrok config add-authtoken 2o5uNsxSgtQgXrgJ3wPgs08qFrw_5qx72Q5rbksc48qVAa3ha


    app_path = Path("/content/SupertCagRag/main.py")  # Update if your app is in a different location
    if not app_path.exists():
        logger.error(f"Error: Could not find {app_path}")
        sys.exit(1)
    log_file = open("/content/logs.txt", "w")  # Open the log file for writing

    try:
        # Run the Streamlit app in the background and write output to log file
        process = subprocess.Popen(["streamlit", "run", str(app_path)],
                                   stdout=log_file, stderr=subprocess.STDOUT)

        # Start ngrok
        ngrok_tunnel = ngrok.connect(PORT)  # Connect to port 8000 (Streamlit's default)
        logger.info('Public URL: %s', ngrok_tunnel.public_url)
        print('Public URL:', ngrok_tunnel.public_url)

        # Keep the script running to prevent ngrok from closing
        input("Press Enter to stop the application...")

        process.terminate()  # Terminate the Streamlit process

    except subprocess.CalledProcessError as e:
        logger.exception(f"Error running Streamlit app: {e}")
        sys.exit(1)
    finally:
        log_file.close()  # Close the log file
        ngrok.kill()  # Terminate ngrok when the script ends

if __name__ == "__main__":
    main()
```
